```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
Dataset=pd.read_csv('cluster.csv')
```

```python
Dataset.head()
```

```python
X=Dataset.iloc[:,[3,4]].values
```

```python
from sklearn.cluster import DBSCAN
DB=DBSCAN(eps=3, min_samples=2).fit(X)
Y_DB=DB.fit_predict(X)
```

```python
Y_DB
```

```python
supervised=pd.DataFrame(Dataset)
```

```python
supervised['Cluster_group']=Y_DB
```

```python
print(supervised)
```

```python
supervised.to_csv("DBSCAN_cluster.csv",index=False)
```

```python
import matplotlib.pyplot as plt
from sklearn.cluster import DBSCAN

# Assuming X is a 2D array with two features
DB = DBSCAN(eps=3, min_samples=2)
Y_DB = DB.fit_predict(X)

# Plotting the points with their assigned cluster labels
plt.figure(figsize=(10, 6))
plt.scatter(X[:, 0], X[:, 1], c=Y_DB, cmap='viridis', marker='o', edgecolor='k')

# Adding labels and title
plt.title('DBSCAN Clustering')
plt.xlabel('Feature 1')
plt.ylabel('Feature 2')

# Show the plot
plt.show()
```
