```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
Dataset=pd.read_csv('cluster.csv')
```

```python
Dataset.head()
```

```python
X=Dataset.iloc[:,[3,4]].values
```

```python
from sklearn.cluster import AffinityPropagation
affinity = AffinityPropagation(random_state=42)
affinity.fit(X)

# Get the cluster centers and labels
cluster_centers_indices = affinity.cluster_centers_indices_
labels = affinity.labels_

# Number of clusters
n_clusters = len(cluster_centers_indices)

print(f"Estimated number of clusters: {n_clusters}")

# Plot the data and cluster centers
plt.scatter(X[:, 0], X[:, 1], c=labels, cmap='viridis', marker='o')
plt.scatter(X[cluster_centers_indices, 0], X[cluster_centers_indices, 1], s=300, c='red', marker='x')
plt.title('Affinity Propagation Clustering')
plt.show()
```

```python
affinity = AffinityPropagation(preference=-50, random_state=42)
affinity.fit(X)
```
