```python
import pandas as pd
```

```python
Dataset=pd.read_csv("Profit Data.csv")
print(Dataset)
```

```python
Dataset.columns
```

```python
Independent=Dataset[['Store_ID', 'Product_ID', 'Units_Sold', 'Unit_Price', 'Total_Sales','Cost_Price', 'Total_Cost']]
print(Independent)
Dependent=Dataset[["Profit"]]
print(Dependent)
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,Y_train,Y_test=train_test_split(Independent,Dependent,test_size=0.30,random_state=0)
```

```python
from sklearn.ensemble import RandomForestClassifier
regressor=RandomForestClassifier(n_estimators=50, criterion='entropy', max_depth=None, min_samples_split=2, min_samples_leaf=1, min_weight_fraction_leaf=0.0, max_features='sqrt', max_leaf_nodes=None, min_impurity_decrease=0.0, bootstrap=True, oob_score=False, n_jobs=None)
regressor.fit(X_train,Y_train)
```

```python
Y_pred=regressor.predict(X_test)
```

```python
from sklearn.metrics import r2_score
r_score=r2_score(Y_test,Y_pred)
print(r_score)
```
