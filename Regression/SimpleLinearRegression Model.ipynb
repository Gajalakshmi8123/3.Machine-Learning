```python
#Step 1: pandas is a library to analyse the data 
# Dataset varaiable is assigned for reading the csv file 
# printing the dataset
import pandas as pd
Dataset=pd.read_csv("Employee_Salary_Detail.csv")
print(Dataset)
```

```python
#step 2: split input/output data
# Independent variable is assigned for Input data
# Dependent variable is assigned for output data
# printing the Independent variable 
# printing the Dependent variable
Independent=Dataset[['Years_of_EXP']]
Dependent=Dataset[['Salary']]
print(Independent)
print(Dependent)
```

```python
#step 3: split train test set data
# importing train_test_split from sklearn.model_selection to splitting the train data and test data
# X_train,X_test,Y_train,Y_test variable assigned for Independent(input variable),Dependent(Output variable),and test_size=0.30 as 30% percentage of the data as test size
# train data as split into 70% and test data as split as 30%
from sklearn.model_selection import train_test_split
x_train,x_test,Y_train,Y_test=train_test_split(Independent,Dependent,test_size=0.30,random_state=0)
```

```python
#step 4: Fit model for (y=wx+b)
# importing the linear regression form sklearn.linear_model
# assigning the regressor variable for LinearRegression() class
# fitting the model for x_train,Y_train
from sklearn.linear_model import LinearRegression
Regressor=LinearRegression()
Regressor.fit(x_train,Y_train)
```

```python
#Step5: finding weight and bias
# finding the weight is between the two points 
# bias is the starting point of the data
Weight=Regressor.coef_
print(Weight)
Bias=Regressor.intercept_
print(Bias)
```

```python
#Step 6: Evalution metrics for predicting the x_test
Y_pred=Regressor.predict(x_test)
```

```python
#Step 7: R_score to find the accuracy of the model
# importing the r2_score from sklearn.metrics 
# printing the R_score to see the model accuracy
from sklearn.metrics import r2_score
R_score=r2_score(Y_test,Y_pred)
print(R_score)
```

```python
#Step 8:Deployment Phase
# saving the model using import pickle
# Filename variable is assigned to set the name of the file "Finalize.regressor.model.sav" 
# dump is to open the file and write binary(wb)

import pickle
Filename="Finalize.regressor.model.sav"
pickle.dump(Regressor,open("Filename","wb"))
```

```python
#step 9: loaded the model for new prediction
loaded_model=pickle.load(open("Finalize_model_linear.sav","rb"))
result=loaded_model.predict([[15]])
print(result)
```
