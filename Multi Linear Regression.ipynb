```python
import pandas as pd
Dataset=pd.read_csv("Performance.csv")
print(Dataset.head(10))
```

```python
Independent=Dataset[["Performance_Rating","Years_of_Experience"]]
Dependent=Dataset[["Salary"]]
print(Independent.head(10))
print(Dependent.head(10))
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,Y_train,Y_test=train_test_split(Independent,Dependent,test_size=0.30,random_state=40)
```

```python
from sklearn.linear_model import LinearRegression
regressor=LinearRegression()
regressor.fit(X_train,Y_train)
```

```python
weight=regressor.coef_
print(weight)
Bias=regressor.intercept_
print(Bias)
```

```python
Y_pred=regressor.predict(X_test)
```

```python
from sklearn.metrics import r2_score
r_score=r2_score(Y_pred,Y_test)
print(r_score)
```

```python
import pickle
Filename="Final_mdel_multilinear.sav"
pickle.dump(regressor,open(Filename,"wb"))
```
