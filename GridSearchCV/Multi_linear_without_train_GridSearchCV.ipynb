```python
import pandas as pd
```

```python
Dataset=pd.read_csv('Insurance_pre.csv')
```

```python
Dataset=pd.get_dummies(Dataset,dtype=int,drop_first=True)
```

```python
Dataset.columns
```

```python
Independent=Dataset[['age', 'bmi', 'children', 'sex_male', 'smoker_yes']]
Dependent=Dataset[[ 'charges']]
```

```python
from sklearn.model_selection import GridSearchCV
from sklearn.linear_model import LinearRegression
Param_grid={'fit_intercept':[True,False],'copy_X':[True,False],'n_jobs':[None],'positive':[True,False]}
```

```python
grid = GridSearchCV(LinearRegression(), Param_grid, refit = True, verbose= 3,n_jobs=-1)
grid.fit(Independent,Dependent)
```

```python
grid_Pred=grid.predict(Independent)
```

```python
re=grid.cv_results_
print("The R_score value for best parameter {}:".format(grid.best_params_))
```

```python
from sklearn.metrics import r2_score
r_score=r2_score(grid_Pred,Dependent)
print(r_score)
```

```python
table=pd.DataFrame.from_dict(re)
print(table)
```
