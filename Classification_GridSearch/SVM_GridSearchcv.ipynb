```python
import pandas as pd
```

```python
Dataset=pd.read_csv('social_network_ads_classification.csv')
```

```python
Dataset.head()
```

```python
Dataset.columns
```

```python
Dataset=pd.get_dummies(Dataset,dtype=int,drop_first=True)
```

```python
Dataset.columns
```

```python
Independent=Dataset[[ 'Age', 'EstimatedSalary', 'Gender_Male']]
Dependent=Dataset[['Purchased']]
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,Y_train,Y_test=train_test_split(Independent,Dependent,test_size=0.30,random_state=0)
```

```python
from sklearn.preprocessing import StandardScaler
scx=StandardScaler()
X_train=scx.fit_transform(X_train)
X_test=scx.transform(X_test)
```

```python
from sklearn.svm import SVC

from sklearn.model_selection import GridSearchCV

Param_grid={'kernel':['linear', 'rbf', 'poly', 'sigmoid'],
            'gamma': ['auto','scale'],
            'C':[10,100,1000,2000,3000]
            }
grid = GridSearchCV(SVC(probability=True), Param_grid, refit = True, verbose = 3,n_jobs=-1,scoring='f1_weighted') 

grid.fit(X_train, Y_train)
```

```python
re=grid.cv_results_
```

```python
grid_Pred=grid.predict(X_test)
```

```python
from sklearn.metrics import confusion_matrix
cm=confusion_matrix(grid_Pred,Y_test)
print(cm)
```

```python
from sklearn.metrics import f1_score
fl_macro=f1_score(grid_Pred,Y_test,average='weighted')
print("The f1 score value for the best parameter{}:".format(grid.best_params_),fl_macro)
```

```python
from sklearn.metrics import classification_report
clf_report=classification_report(grid_Pred,Y_test)
print(clf_report)
```

```python
from sklearn.metrics import roc_auc_score
roc_auc_score(Y_test,grid.predict_proba(X_test)[:,1])
```
